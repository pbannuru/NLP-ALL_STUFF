```python
def sliding_window(txt):
    N = len(txt)-1
    for i in range(N):
        yield txt[i], txt[i+1]  # return pair current input, target value
    
```

```python
import itertools as it
```

# Using Python itertools.chain() to chain iterables together
## The Python itertools.chain() method generates an iterator from multiple iterables.

### This simply chains all the iterables together into one sequence and returns a single iterator to that combined sequence.

#### The syntax for this method is as follows

# iterator = itertools.chain(*sequence)

```python
for txt in headlines[:10000]:
    print (txt,len(txt)-1)
```

```python
window=list(it.chain(*[sliding_window(txt) for txt in headlines[:10000]]))
window
```

```python
pd.DataFrame(window)
```

```python
unique_letters=pd.DataFrame(window)[0].unique()
unique_letters
```

```python
mapping={char: idx for idx,char in enumerate(unique_letters)}
mapping
```

```python
[mapping[w[0]] for w in window]
```

```python
integer_in =np.array([mapping[w[0]] for w in window])
integer_in
```

```python
integer_out =np.array([mapping[w[1]] for w in window])
integer_out
```

```python
num_letters = len(mapping)
num_letters
```

```python
embedding = tf.keras.layers.Embedding(num_letters,2,input_length=1)
model_emb = tf.keras.Sequential(embedding)
```

```python
out_array = model_emb.predict(integer_in)
```

```python
out_array.shape
```

```python
idx_of_char=np.array([list(mapping.values())]).T
```

```python
idx_of_char.shape
```

```python
translator = {v:k for k,v in mapping.items()}
translator
```

```python
preds = model_emb.predict(idx_of_char) 
preds
```

```python
plt.scatter(preds[:,0,0],preds[:,0,1])
for i,idx in enumerate(idx_of_char):
    plt.text(preds[i,0,0],preds[i,0,1],translator[idx[0]]
plt.show()
```

```python
plt.scatter(preds[:,0,0],preds[:,0,1])
```

```python
LAYERS=[
        embedding,
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(num_letters,activation="softmax")
]



model_pred = tf.keras.Sequential(LAYERS)
```

```python
model_pred.summary()
```

```python
adam = tf.keras.optimizers.Adam(learning_rate=0.001)
model_pred.compile(adam,"categorical_crossentropy",metrics=["accuracy"])
```
