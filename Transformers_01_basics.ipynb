```python
classifeir = pipeline('sentiment-analysis')
```

```python
classifeir('I love my country')
```

```python
classifier_zero_shot =pipeline('zero-shot-classification')
```

```python
classifier_zero_shot (
    "this course is about Artificial Intelligence",
    candidate_labels=['politics','business','technology','education']
)
```

```python
classifier_zero_shot (
    "this course is about Artificial Intelligence in an institution,here we learn theory and practicals",
    candidate_labels=['politics','business','education']
)
```
